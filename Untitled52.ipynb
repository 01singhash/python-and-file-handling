```python
import csv
from IPython.display import display, HTML
import pandas as pd

# List to store student records
student_records = []

# Function to add a student
def add_student(name, roll_number, marks):
    if len(marks) != 3:
        print("‚ùå Please enter exactly 3 marks.")
        return

    student = {
        "Name": name,
        "Roll Number": roll_number,
        "Marks": marks
    }
    student_records.append(student)
    print("‚úÖ Student added successfully!")

# Function to display all records
def display_records():
    if not student_records:
        print("üì≠ No records to display.")
        return

    rows = []
    for student in student_records:
        row = {
            "Name": student["Name"],
            "Roll Number": student["Roll Number"],
            "Subject 1": student["Marks"][0],
            "Subject 2": student["Marks"][1],
            "Subject 3": student["Marks"][2]
        }
        rows.append(row)

    df = pd.DataFrame(rows)
    display(df)

# Function to calculate average marks
def calculate_averages():
    if not student_records:
        print("üì≠ No records to calculate.")
        return

    averages = []
    for student in student_records:
        avg = sum(student["Marks"]) / 3
        averages.append({
            "Name": student["Name"],
            "Roll Number": student["Roll Number"],
            "Average Marks": round(avg, 2)
        })

    df = pd.DataFrame(averages)
    display(df)

# Function to save to CSV
def save_to_csv(filename="student_records.csv"):
    if not student_records:
        print("üì≠ No records to save.")
        return

    try:
        with open(filename, mode="w", newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Name", "Roll Number", "Subject 1", "Subject 2", "Subject 3"])
            for student in student_records:
                writer.writerow([student["Name"], student["Roll Number"], *student["Marks"]])
        print(f"üíæ Records saved to {filename} successfully!")

        # Create download link
        from google.colab import files
        files.download(filename)

    except Exception as e:
        print(f"‚ùå Error saving file: {e}")
```

```python
# Example: Add student
add_student("Alice", "101", [85, 90, 95])
add_student("Bob", "102", [78, 82, 88])
```

```python
# Show all student records
display_records()
```

```python
# Calculate average marks
calculate_averages()
```

```python
# Save records to CSV and download
save_to_csv()

```
